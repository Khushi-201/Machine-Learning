```python
import numpy as np
```

```python
X=iris['data'][:,3:]
print(X)
y=(iris["target"]==2).astype(np.int_)  #binary classifier
```

```python
from sklearn.linear_model import LogisticRegression 
clf=LogisticRegression()
clf.fit(X,y)
```

```python
example=clf.predict(([[1.6]]))
print(example)
```

```python
import matplotlib.pyplot as plt
X_new=np.linspace(0,3,1000).reshape(-1,1) #1000 values of petal width between 0 and 3, reshape(-1,1) make it as a array
y_prob=clf.predict_proba(X_new)
plt.plot(X_new,y_prob[:,1],label='verginica')
```
